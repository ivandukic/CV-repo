```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Importing parallax zero-point offset code 
# (https://gitlab.com/icc-ub/public/gaiadr3_zeropoint)
from zero_point import zpt 
zpt.load_tables()

pd.set_option('display.max_columns', None)
```

```python
# Modify this cell depending on your OS and table name. 
separator = '/' 
table_name = 'asassn_miras_join_gedr3_xm_2MASS.csv'
```

```python
# Taking the absolute path of the current working
# directory and its parental directory
absolutepath = os.path.abspath("")
parentDirectory = os.path.dirname(absolutepath)
dataDirectory = parentDirectory + separator + 'data'
```

```python
#Reading data
stars = pd.read_csv(dataDirectory + separator + table_name)
```

```python
#Dropping NAN values
stars_no_nan = stars.dropna(subset=['phot_rp_mean_mag_gedr3',
                                    'bp_rp_gedr3','parallax_gedr3',
                                    'Jmag','Kmag'])
```

# Zero-point calculation

```python
gmag = stars_no_nan['phot_g_mean_mag_gedr3'].values
nueffused = stars_no_nan['nu_eff_used_in_astrometry_gedr3'].values
psc = stars_no_nan['pseudocolour_gedr3'].values
ecl_lat = stars_no_nan['ecl_lat_gedr3'].values
soltype = stars_no_nan['astrometric_params_solved_gedr3'].values
```

## Blindly use the get_zpt function

```python
# Using the `get_zpt()` function blindly will fail if there are
# source with 2-p solutions in the input list. 
#  Thus the line of code below will lead to a runtime error.
zpvals = zpt.get_zpt(gmag, nueffused, 
                     psc, ecl_lat, soltype)
```

## Remove the 2-p solutions

```python
# Re-running the line of above with the 2-p solution removed works.
# The code automatically deals with a mix of 5-p and 6-p solutions.
valid = soltype>3
zpvals = zpt.get_zpt(gmag[valid], nueffused[valid], 
                     psc[valid], ecl_lat[valid], 
                     soltype[valid])
```

As explained in Lindegren et al. 2020, the interpolations are only calibrated within the following intervals:

1. $G$ magnitude: 
    6 < phot_g_mean_mag < 21
2. Colour:
    1. 1.1 < nu_eff_used_in_astrometry < 1.9 (5-p sources)
    2. 1.24 < pseudocolour < 1.72 (6-p sources)
        
Outside these ranges, the zero-point obtained is an **extrapolation**.

```python
# This makes the code return NaN for the sources that 
# fall outside the recomended limits,
# instead of an extrapolated zero point
valid = soltype>3
zpvals = zpt.get_zpt(gmag[valid], nueffused[valid], 
                     psc[valid], ecl_lat[valid], 
                     soltype[valid],_warnings=False)
```

```python
# Converting NaN values to 0.0. 
np.nan_to_num(x=zpvals, copy=False, nan=0.0)
```

# ABL method

```python
stars_valid = stars_no_nan[valid]
```

```python
global_zpo = -0.014
parallax = stars_valid['parallax_gedr3'] - zpvals - global_zpo
```

```python
#Wesenheit index:
w_jk = stars_valid['Kmag'] - 0.686 * (stars_valid['Jmag'] \
                                      - stars_valid['Kmag'])
```

```python
abl = parallax * pow(10, 0.2*w_jk-2)
logP = np.log10(stars_valid['period'])
```

```python
# Parallax total standard error calculation.
zpo_err_individual = 0.013
global_zpo_err = 0.06

# We increased parallax error from edr3 by 10%.
std_err_edr3 = np.sqrt((1.1 * stars_valid['parallax_error_gedr3'])**2 
                       + zpo_err_individual**2 + global_zpo_err**2)
```

```python
# Model function
def func(x, alpha, beta):
    return pow(10,0.2 * (alpha*x + beta))
```

```python
popt, pcov = curve_fit(f=func, xdata = logP,
                       ydata = abl, sigma=std_err_edr3)

perr = np.sqrt(np.diag(pcov))
```

```python
print('W = alpha*logP + beta')
print('-----------------------')
print('alpha = {} ± {}'.format(popt[0],perr[0]))
print('beta = {} ± {}'.format(popt[1],perr[1]))
```

```python
xs = np.linspace(1.8, 3, 100)
plt.figure(figsize=[14,8])
plt.scatter(logP, abl, s=1)

plt.plot(xs, func(xs, popt[0], popt[1]),
         color='red',
         label='10^(0.2*(alpha*log(P) + beta))')

plt.title("w_jk", fontsize = 16)
plt.xlabel("log(P)", fontsize = 16)
plt.ylabel("ABL", fontsize = 16)
plt.gca().invert_yaxis()

plt.text(x=2.05, y=0.6,
         s="""W = alpha * log(P) + beta \n 
         alpha = {} ± {} \n beta = {} ± {}""".format(popt[0],perr[0],popt[1],perr[1]),
         ha="center", va="center", size = 12)

plt.legend()
```

# C/O-rich diagram
From: Lebzelter et al.: Classification of LPVs using Gaia and 2MASS

```python
# Reading data for C/O-rich diagram
COdiag = pd.read_csv(dataDirectory + separator + 'gdr2_LMC_LPVs_join_2MASSv2.csv')
```

```python
COdiag_W_jk = COdiag['ks_m'] - 0.686*(COdiag['j_m']\
                                     -COdiag['ks_m'])

COdiag_W_bprp = COdiag['phot_rp_mean_mag'] \
                -1.3 * (COdiag['phot_bp_mean_mag'] \
                        -COdiag['phot_rp_mean_mag'])

w_bprp = stars_valid['phot_rp_mean_mag_gedr3']\
         - 1.3 * stars_valid['bp_rp_gedr3']

abs_w_jk = w_jk + 5 * (np.log10(parallax/1000) + 1) 
abs_w_bprp = w_bprp + 5 * (np.log10(parallax/1000) + 1) 
abs_k = stars_valid['Kmag'] + 5 * (np.log10(parallax/1000) + 1) 
```

```python
# Note that we added distance modulus to the LMC (18.5 mag)
Orich_filt = ((abs_w_bprp - abs_w_jk) <= (0.7 + 0.15*(abs_k + 18.5 -10.8)**2))
Crich_filt = ((abs_w_bprp - abs_w_jk) > (0.7 + 0.15*(abs_k + 18.5 -10.8)**2))
```

```python
plt.figure(figsize = [12,8])

plt.scatter(COdiag_W_bprp - COdiag_W_jk, COdiag['ks_m'], 
            c = COdiag['j_m'] - COdiag['ks_m'],
            cmap='jet', vmin=1, vmax=2 ,s=1)

plt.scatter(abs_w_bprp[Orich_filt] - abs_w_jk[Orich_filt],
            abs_k[Orich_filt] + 18.5, 
            s= 20, marker='^', color='black')

plt.scatter(abs_w_bprp[Crich_filt] - abs_w_jk[Crich_filt], 
            abs_k[Crich_filt] + 18.5, 
            s= 20, marker='^', color='red')

plt.colorbar(label = "$J-K_S$", orientation="vertical",)

plt.xlim(-2,4)
plt.ylim(6, 14)
plt.ylabel('$K_S$', fontsize = 16)
plt.xlabel('$W_{bprp} - W_{jk}$', fontsize =16)
plt.gca().invert_yaxis()
```

# ABL with C/O-rich classification

## O-rich 

```python
popt_O, pcov_O = curve_fit(f=func, 
                           xdata=logP[Orich_filt],
                           ydata=abl[Orich_filt], 
                           sigma=std_err_edr3[Orich_filt])

perr_O = np.sqrt(np.diag(pcov_O))
```

```python
print('O-rich')
print('W = alpha*logP + beta')
print('-----------------------')
print('alpha = {} ± {}'.format(popt_O[0],perr_O[0]))
print('beta = {} ± {}'.format(popt_O[1],perr_O[1]))
```

## C-rich

```python
popt_C, pcov_C = curve_fit(f=func,
                           xdata=logP[Crich_filt],
                           ydata=abl[Crich_filt],
                           sigma=std_err_edr3[Crich_filt])

perr_C = np.sqrt(np.diag(pcov_C))
```

```python
print('C-rich')
print('W = alpha*logP + beta')
print('-----------------------')
print('alpha = {} ± {}'.format(popt_C[0],perr_C[0]))
print('beta = {} ± {}'.format(popt_C[1],perr_C[1]))
```

## C/O-rich PLR plot

```python
xs = np.linspace(1.8, 3, 100)
plt.figure(figsize=[14,8])

plt.scatter(logP[Orich_filt], abl[Orich_filt],
            s=10, color='black', label='O-rich')

plt.plot(xs, func(xs, popt_O[0], popt_O[1]),
         color='blue', label = 'O-rich fit',
         linewidth = 3)

plt.scatter(logP[Crich_filt], abl[Crich_filt],
            s=10, color='red', label='C-rich')

plt.plot(xs, func(xs, popt_C[0], popt_C[1]),
         color='red', label='C-rich fit',
         linewidth = 3)

plt.ylim(-0.5, 0.5)
plt.title("w_jk", fontsize = 20)
plt.xlabel("log(P)", fontsize = 16)
plt.ylabel("ABL", fontsize = 16)
plt.gca().invert_yaxis()

plt.text(x=2.05, y=-0.3,
         s="""W = alpha * log(P) + beta \n
         O-rich \n
         alpha = {} ± {} \n beta = {} ± {} \n
         C-rich \n
         alpha = {} ± {} \n beta = {} ± {}""".format(popt_O[0],perr_O[0],
                                                     popt_O[1],perr_O[1],
                                                     popt_C[0],perr_C[0],
                                                     popt_C[1],perr_C[1]),
         ha="center", va="center", size = 12)

plt.legend()
```

# ABL method for RUWE < 1.4 and 6 < G mag< 19

```python
# First, we will check how many stars is from LMC
lmc_filt = (stars_valid['ra_gedr3'] > 50) & (stars_valid['ra_gedr3'] < 105)\
           & (stars_valid['dec_gedr3'] > -77) & (stars_valid['ra_gedr3'] < -61)\
           & (stars_valid['pmra_gedr3'] > 1.2) & (stars_valid['pmra_gedr3'] < 2.5)\
           & (stars_valid['pmdec_gedr3'] > -0.8) & (stars_valid['pmdec_gedr3'] < 1.5)\
           & (stars_valid['parallax_gedr3'] < 0.5)

lmc_filt.value_counts()
```

As we can see there is no stars from Large Magellanic Cloud.

```python
ruwe_and_g_mag_filter = (stars_valid['ruwe_gedr3'] < 1.4)\
                        & (stars_valid['phot_g_mean_mag'] > 6)\
                        & (stars_valid['phot_g_mean_mag'] < 19)
```

```python
popt_filtered_data, pcov_filtered_data = curve_fit(f=func,
                                                   xdata=logP[ruwe_and_g_mag_filter],
                                                   ydata=abl[ruwe_and_g_mag_filter],
                                                   sigma=std_err_edr3[ruwe_and_g_mag_filter])

perr_filtered_data = np.sqrt(np.diag(pcov_filtered_data))
```

```python
print('W = alpha*logP + beta')
print('-----------------------')
print('alpha = {} ± {}'.format(popt_filtered_data[0],perr_filtered_data[0]))
print('beta = {} ± {}'.format(popt_filtered_data[1],perr_filtered_data[1]))
```
