```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

pd.set_option('display.max_columns', None)
```

```python
# The purpose of this cell is to make code runs smoothly 
# on other machines without explicitly modifying the
# absolute path to the csv files.
# Taking the absolute path of the current working
# directory and its parental directory

separator='/'
absolutepath = os.path.abspath("")
parentDirectory = os.path.dirname(absolutepath)
dataDirectory = parentDirectory + separator + 'data'
```

```python
# Reading table from local PC. We will keep just Mira stars from OGLE data.
bulge_data = pd.read_csv(dataDirectory + separator + "blg_xm_final.csv")
ogle_miras = bulge_data[bulge_data['type_ogle'] == 'Mira']

asassn_miras = pd.read_csv(dataDirectory + separator \
                           +'asassn_miras_join_gedr3.csv')
```

```python
ogle_miras
```

```python
ogle_miras[['inty_mean_I_ogle', 'inty_mean_V_ogle',
            'parallax_over_error']].dtypes
```

```python
# Replacing '-' values to 'Nan' so that we can convert 'inty_mean_V_ogle' column to float64
ogle_miras['inty_mean_V_ogle'].replace(to_replace = '-',
                                       value = 'Nan', inplace = True)

ogle_miras = ogle_miras.astype(dtype={'inty_mean_V_ogle' : 'float64'}).copy()
```

```python
asassn_miras[['mean_vmag', 'parallax_over_error']].dtypes
```

```python
print('Number of Miras without OGLE mean I-band value: {}'.format(len(ogle_miras['inty_mean_I_ogle']) - ogle_miras['inty_mean_I_ogle'].isnull().value_counts()[0]))
print('Number of Miras without OGLE mean V-band value: {}'.format(len(ogle_miras['inty_mean_V_ogle']) - ogle_miras['inty_mean_V_ogle'].isnull().value_counts()[0]))
print('Number of Miras without OGLE parallax_over_error value: {}'.format(len(ogle_miras['parallax_over_error']) - ogle_miras['parallax_over_error'].isnull().value_counts()[0]))
print()
print('There is no ASAS-SN I magnitude !!!')
print('Number of Miras without ASAS-SN mean V-band value: {}'.format(len(asassn_miras['mean_vmag']) - asassn_miras['mean_vmag'].isnull().value_counts()[0]))
print('Number of Miras without ASAS-SN parallax_over_error value: {}'.format(len(asassn_miras['parallax_over_error']) - asassn_miras['parallax_over_error'].isnull().value_counts()[0]))
```

```python
plt.figure(figsize=[16,9])

plt.ylabel('Number of Miras', fontsize = 16)
plt.xlabel('Mean I magnitude', fontsize = 16)
plt.xticks(list(np.linspace(10, 20, 11)))
plt.grid()
plt.xlim(10,20)

plt.hist(ogle_miras['inty_mean_I_ogle'], 
         bins = list(np.linspace(10, 20, 101)),
         histtype = 'step', linewidth = 3, 
         label = 'OGLE')

plt.legend()
```

```python
plt.figure(figsize=[16,9])

plt.ylabel('Number of Miras', fontsize = 16)
plt.xlabel('Mean V magnitude', fontsize = 16)
plt.xticks(list(np.linspace(7, 23, 17)))
plt.grid()
plt.xlim(7,23)

plt.hist(ogle_miras['inty_mean_V_ogle'], 
         bins = list(np.linspace(11, 23, 121)),
        histtype = 'step', linewidth = 3, 
         label = 'OGLE', color = 'blue')

plt.hist(asassn_miras['mean_vmag'],
         bins = list(np.linspace(7, 23, 161)),
         histtype = 'step', linewidth = 3,
         label = 'ASAS-SN', color = 'red')

plt.legend()
```

```python
plt.figure(figsize=[16,9])

plt.xlabel('Mean V magnitude', fontsize = 16)
plt.ylabel('Parallax error (%)', fontsize = 16)
plt.title ('ASAS-SN (gdr2)', fontsize = 20)

plt.hlines(10, 8, 18, color = 'red',
           linewidth = 3, linestyle = '--')

plt.hist2d(asassn_miras['mean_vmag'],
           1/asassn_miras['parallax_over_error'] * 100,
           bins = 80, cmin = 1,
           range = [[8,18], [0, 100]])

plt.colorbar()
```

```python
plt.figure(figsize=[16,9])

plt.xlabel('Mean V magnitude', fontsize = 16)
plt.ylabel('Parallax error (%)', fontsize = 16)
plt.title ('OGLE', fontsize = 20)

plt.hlines(10, 4, 23, color = 'red',
           linewidth = 3, linestyle = '--')

plt.hist2d(ogle_miras['inty_mean_V_ogle'],
           1/ogle_miras['parallax_over_error'] * 100,
           bins = 80, cmin = 1,
           range = [[14,23], [0, 100]])

plt.colorbar()
```

```python
plt.figure(figsize=[16,9])
plt.xticks(list(np.linspace(0, 100, 11)))
plt.xlim(0, 100)
plt.grid()

plt.title('Number of stars in each parallax error interval',
          fontsize = 16)

plt.ylabel('Number of stars', fontsize = 16)
plt.xlabel('Relative parallax error (%)', fontsize = 16)

plt.hist(1/ogle_miras['parallax_over_error']*100, 
         bins = list(np.linspace(0, 100, 11)),
         histtype = 'step', linewidth = 3,
         label = 'OGLE', color = 'blue')

plt.hist(1/asassn_miras['parallax_over_error']*100,
         bins = list(np.linspace(0, 100, 11)),
         histtype = 'step', linewidth = 3,
         label = 'ASAS-SN', color = 'red')

plt.hist(1/asassn_miras['parallax_over_error_gedr3']*100,
         bins = list(np.linspace(0, 100, 11)),
         histtype = 'step', linewidth = 2,
         label = 'ASAS-SN (edr3)', color = 'yellow',
         linestyle = '--')

plt.legend()
```

```python
plt.figure(figsize=[16,9])

plt.xlabel('Mean V magnitude', fontsize = 16)
plt.ylabel('Parallax error (%)', fontsize = 16)
plt.title ('ASAS-SN (edr3)', fontsize = 20)

plt.hlines(10, 8, 18, color = 'red',
           linewidth = 3, linestyle = '--')

plt.hist2d(asassn_miras['mean_vmag'], 
           1/asassn_miras['parallax_over_error_gedr3'] * 100, 
           bins = 80, cmin = 1, range = [[8,18], [0, 100]])

plt.colorbar()
```
